```python
import urllib.request
from bs4 import BeautifulSoup
```

```python
url = "https://www.jma.go.jp/jp/quake/20190113093158493-13182843.html?fbclid=IwAR1oF---cpOkApf1bRBoAw17YnYSeTPVznmG81_jd7S5EOVOBPU3HfzIabo"
```

```python
html = urllib.request.urlopen(url)
```

```python
soup = BeautifulSoup(html, "html.parser")
```

```python
soup
```

```python
table = soup.find("table",class_="textframe")
```

```python
information = []

for text in table.find_all("td"):
    if len(text)>10:
        information.append(text)
    else:
        print("aa")
```

```python
information
```

```python
elems = soup.select('a')
for elem in elems:
    print('{} ({})'.format(elem.getText(), elem.get('href')))
```
